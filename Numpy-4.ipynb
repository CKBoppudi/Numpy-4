# Very important example.

```python
A = np.arange(1,10).reshape(1,9)
print(A)
```

```python
B = A[:,::2] # Accesing all the rows and all the columns with a "stride of 2"
print(B)
```

```python
A = np.arange(1,10).reshape(1,9)
B = A.copy()
print(A)
print(B)
```

```python
A[0] = 100
print(A)
print(B)
```

```python
A = np.arange(1,11)
B = np.split(A , 2)
A = A * 2
print(A)
print(B)
```

```python
np.shares_memory(A,B)
```
